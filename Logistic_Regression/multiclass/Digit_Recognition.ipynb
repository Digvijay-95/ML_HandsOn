```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_digits
```

```python
digits=load_digits()
```

```python
dir(digits)
```

```python
len(digits.data[0])
```

```python
print(digits.data[0])
```

```python
import matplotlib.pyplot as plt
```

```python
plt.gray()
for i in range(10):
    plt.matshow(digits.images[i])
```

```python
digits.target[1:9]
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(digits.data, digits.target,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

```python
plt.matshow(digits.images[69])
```

```python
digits.target[69]
```

```python
model.predict([digits.data[69]])
```

```python
ypredict=model.predict(xtest)
from sklearn.metrics import confusion_matrix
```

```python
matrix=confusion_matrix(ytest,ypredict)
```

```python
matrix
```

```python
import seaborn as sb
```

```python
plt.figure(figsize=(10,6))
sb.heatmap(matrix,annot=True)
plt.xlabel=('prediction')
plt.ylabel=('Truth')
```
